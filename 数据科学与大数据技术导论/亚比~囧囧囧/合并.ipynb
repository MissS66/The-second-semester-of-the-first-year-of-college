```python
import pandas as pd  
  
# 假设 df1 和 df2 分别是你的两个 DataFrame  
# ...  
df1 = pd.read_csv("车辆全国月降价信息.csv", encoding='utf-8')
df2=pd.read_csv('车辆口碑排行信息.csv',encoding='utf-8')  
df3=pd.read_csv('车辆全国月关注信息.csv',encoding='utf-8')  
df4=pd.read_csv('车辆销售排行信息.csv',encoding='utf-8',nrows=236)  
# 创建一个映射，其中键是'车名'，值是'评分'  
score_map = df2.set_index('车名')['评分'].to_dict()  
score_map2=df3.set_index('车名')['全国月关注'].to_dict()  
score_map3=df4.set_index('车名')['全国月销量'].to_dict()  
# 检查是否有任何车名在df1中但不在df2的映射中  
# missing_car_names = set(df1['车名']) - set(score_map.keys())  

# 使用映射来更新df1中的评分列  
# 如果df1中没有评分列，先创建一个全NaN的列  
if '全国月关注' not in df1.columns:  
    df1['全国月关注'] = pd.NA  # 或者使用 np.nan，但 pd.NA 在处理缺失数据时可能更直观  
  
df1['全国月关注'] = df1['车名'].map(score_map2)  
if '评分' not in df1.columns:  
    df1['评分'] = pd.NA  # 或者使用 np.nan，但 pd.NA 在处理缺失数据时可能更直观  
  
df1['评分'] = df1['车名'].map(score_map) 
if '全国月销量' not in df1.columns:  
    df1['全国月销量'] = pd.NA  # 或者使用 np.nan，但 pd.NA 在处理缺失数据时可能更直观  
  
df1['全国月销量'] = df1['车名'].map(score_map3)  
#保存csv
 


#保存csv

df1.to_excel('合并.xlsx',index=False)  
# 现在 df1 中包含了从 df2 映射过来的评分（如果有的话）  
print(df1)
```

```python
import pandas as pd  
  
# 读取CSV文件  
df1 = pd.read_csv("车辆全国月降价信息.csv", encoding='utf-8')  
df2 = pd.read_csv('车辆口碑排行信息.csv', encoding='utf-8')  
  
# 清洗数据：转换为小写并去除前后空格  
df1['车名'] = df1['车名'].str.lower().str.strip()  
df2['车名'] = df2['车名'].str.lower().str.strip()  
  
# 使用 outer join 合并两个 DataFrame  
merged_df = pd.merge(df1, df2, on='车名', how='left')  
#保存csv
merged_df.to_csv('合并后.csv', encoding='gbk', index=False)  
# 打印合并后的 DataFrame  
print(merged_df)
```
